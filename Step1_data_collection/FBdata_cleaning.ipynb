```python
# CLEAN THE INTERESTS
for index, row in df.iterrows():
    txt = row['interests'].strip('}{u').replace('u', '').split(', ')
    name = txt[-1].split(': ')[-1].strip("'")
    noin = len(txt[:-1])
    interests = []
    for i, inter in enumerate(txt[:-1]):
        if i==0:
            inter = inter.split(': ')[-1].strip('[')
        elif i==noin-1:
            inter = inter.strip(']')
        interests.append(inter)
        
    df.loc[index, 'interest_name'] = name
    df.loc[index, 'interest_ids'] = ' '.join(interests)
```
